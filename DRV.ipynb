```python
from scipy.stats import binom, poisson, geom, hypergeom
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.set_printoptions(precision=4)
pd.options.display.float_format = '{:.4f}'.format

def mode(xs, pmf):
    max_p = np.max(pmf)
    indices = np.where(np.isclose(pmf, max_p))[0]
    return xs[indices] 

def draw(xs, pmf, name):
    plt.figure(figsize=(6, 4))
    plt.bar(xs, pmf, align='center', alpha=0.7)
    plt.plot(xs, pmf, marker='o', linestyle='-')
    plt.title(f"{name} - многоугольник распределения")
    plt.xlabel("x")
    plt.ylabel("P(X = x)")
    plt.grid(axis='y', linestyle=':', linewidth=0.5)
    plt.tight_layout()
    plt.show()
```

#### В мастерской проверяют партию из 10 деталей на исправность. Вероятность того, что деталь прошла тест на исправность - 0.3. Найти математическое ожидание, дисперсию, среднеквадратичное отклонение и моду данной случайной величины. Построить многоугольник распределения

```python
n, p = 10, 0.3

xs = np.arange(0, n + 1)

binomial_dist = binom(n, p)

pmf = binomial_dist.pmf(xs)

df = pd.DataFrame(data=pmf.reshape(1, -1), columns=xs, index=["p"])

display(df)

print(f"Мат. ожидание = {binomial_dist.mean():.4f}")
print(f"Дисперсия = {binomial_dist.var():.4f}")
print(f"Ст. отклонение = {binomial_dist.std():.4f}")
print(f"Мода = {mode(xs, pmf)}")

draw(xs, pmf, "Биномиальное распределение")
```

#### В течении минуты в отдел технической поддержки клиентов фирмы поступает в среднем 2.5 запроса. Найти математическое ожидание, дисперсию, среднеквадратичное отклонение и моду данной случайной величины. Построить многоугольник распределения

```python
lmb = 2.5

xs = np.arange(0, 11)

poisson_dist = poisson(lmb)

pmf = poisson_dist.pmf(xs)

df = pd.DataFrame(data=pmf.reshape(1, -1), columns=xs, index=["p"])

display(df)

print(f"Мат. ожидание = {poisson_dist.mean():.4f}")
print(f"Дисперсия = {poisson_dist.var():.4f}")
print(f"Ст. отклонение = {poisson_dist.std():.4f}")
print(f"Мода = {mode(xs, pmf)}")

draw(xs, pmf, "Распределение Пуассона")
```

#### Сборочная линия процессоров работает до первого неисправного процессора. Вероятность неисправности процессора равно 0.2. Найти математическое ожидание, дисперсию, среднеквадратичное отклонение и моду данной случайной величины. Построить многоугольник распределения

```python
p = 0.2

xs = np.arange(1, 16)

geom_dist = geom(p)

pmf = geom_dist.pmf(xs)

df = pd.DataFrame(data=pmf.reshape(1, -1), columns=xs, index=["p"])

display(df)

print(f"Мат. ожидание = {geom_dist.mean():.4f}")
print(f"Дисперсия = {geom_dist.var():.4f}")
print(f"Ст. отклонение = {geom_dist.std():.4f}")
print(f"Мода = {mode(xs, pmf)}")

draw(xs, pmf, "Геометрический распределение")
```

#### Поставщик поставил 50 деталей, среди которых 8 оказались дефектными. Случайным образом выбираются 6 деталей из всех поставленных. Составить закон распределения числа дефектных деталей среди выбранных. Найти математическое ожидание, дисперсию, среднеквадратичное отклонение и моду данной случайной величины. Построить многоугольник распределения

```python
N, M, n = 50, 8, 6

xs = np.arange(0, n + 1)

hypergeom_dist = hypergeom(N, M, n)

pmf = hypergeom_dist.pmf(xs)

df = pd.DataFrame(data=pmf.reshape(1, -1), columns=xs, index=["p"])

display(df)

print(f"Мат. ожидание = {hypergeom_dist.mean():.4f}")
print(f"Дисперсия = {hypergeom_dist.var():.4f}")
print(f"Ст. отклонение = {hypergeom_dist.std():.4f}")
print(f"Мода = {mode(xs, pmf)}")

draw(xs, pmf, "Гипергеометрическое распределение")
```
