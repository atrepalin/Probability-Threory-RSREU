```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

display(
    Markdown(
        "### Датасет содержит признаки пациентов (возраст, пол, давление, ЭКГ и др.),"
        "а также бинарную целевую переменную - наличие болезни сердца."
    )
)

df: pd.DataFrame = kagglehub.dataset_load(
    KaggleDatasetAdapter.PANDAS,
    "rishidamarla/heart-disease-prediction",
    "Heart_Disease_Prediction.csv",
)

df["Heart Disease"] = df["Heart Disease"].map({"Presence": 1, "Absence": 0})

df = df.rename(
    columns={
        "Age": "Возраст",
        "Sex": "Пол",
        "Chest pain type": "Тип боли в груди",
        "BP": "Артериальное давление",
        "Cholesterol": "Холестерин",
        "FBS over 120": "Сахар натощак >120",
        "EKG results": "Результаты ЭКГ",
        "Max HR": "Макс. ЧСС",
        "Exercise angina": "Стенокардия при нагрузке",
        "ST depression": "Депрессия сегмента ST",
        "Slope of ST": "Наклон сегмента ST",
        "Number of vessels fluro": "Кол-во сосудов (флюро)",
        "Thallium": "Таллий-тест",
        "Heart Disease": "Болезнь сердца",
    }
)

print("Размер датасета:", df.shape)
display(df.head())
display(df.info())
display(df.describe())


def conditional_probs(data: pd.DataFrame, col):
    return (
        data.groupby(col, observed=True)["Болезнь сердца"]
        .mean()
        .reset_index(name=f"P(Болезнь сердца|{col})")
        .style.hide(axis="index")
    )

def bayes_conditional(data: pd.DataFrame, col):
    p_target1 = (data["Болезнь сердца"] == 1).mean()
    p_target0 = 1 - p_target1

    rows = []
    for val in data[col].dropna().unique():
        p_val = (data[col] == val).mean() 

        p_target1_given_val = (
            data.loc[data[col] == val, "Болезнь сердца"].mean()
        )
        p_target0_given_val = 1 - p_target1_given_val

        p_val_given_target1 = (p_target1_given_val * p_val) / p_target1
        p_val_given_target0 = (p_target0_given_val * p_val) / p_target0

        rows.append([val, p_val_given_target1, p_val_given_target0])

    return pd.DataFrame(
        rows,
        columns=[col, f"P({col}|Болезнь сердца=1)", f"P({col}|Болезнь сердца=0)"],
    ).style.hide(axis="index")

display(
    Markdown(
        "### Непрерывные признаки (возраст, давление, ЧСС, холестерин) разбиваем на интервалы (бины),"
        "чтобы можно было считать вероятности для категорий."
    )
)

df_grouped = pd.DataFrame(df)

df_grouped["Пол"] = df["Пол"].map({0: "Женщина", 1: "Мужчина"})

df_grouped["Артериальное давление"] = pd.cut(
    df["Артериальное давление"],
    bins=[80, 120, 140, 160, 200],
    labels=["<=120", "121-139", "140-159", "160+"],
)
df_grouped["Возраст"] = pd.cut(
    df["Возраст"], bins=[20, 40, 55, 70, 100], labels=["20-39", "40-54", "55-69", "70+"]
)
df_grouped["Макс. ЧСС"] = pd.cut(
    df["Макс. ЧСС"],
    bins=[70, 100, 130, 160, 200],
    labels=["70-99", "100-129", "130-159", "160+"],
)
df_grouped["Холестерин"] = pd.cut(
    df["Холестерин"],
    bins=[100, 200, 250, 300, 400, 600],
    labels=["<=200", "201-250", "251-300", "301-400", "401+"],
)

display(df_grouped.head())

display(
    Markdown(
        "### Значения показывают условные вероятности $P(\\text{Болезнь сердца} | X)$ для разных категорий."
    )
)

display(Markdown("### Вероятность болезни по возрастным группам"))
display(conditional_probs(df_grouped, "Возраст"))

display(Markdown("### Вероятность болезни по полу"))
display(conditional_probs(df_grouped, "Пол"))

display(Markdown("### Вероятность болезни по возрастным максимальной ЧСС"))
display(conditional_probs(df_grouped, "Макс. ЧСС"))

display(Markdown("### Вероятность болезни по наличию стенокардии при нагрузке"))
display(conditional_probs(df_grouped, "Стенокардия при нагрузке"))

display(Markdown("### Вероятность болезни по холестерину"))
display(conditional_probs(df_grouped, "Холестерин"))

display(
    Markdown(
        "### Значения показывают условные вероятности $P(X | \\text{Болезнь сердца})$ для разных категорий по формуле Байеса."
    )
)

display(Markdown("### Вероятность пола по формуле Байеса"))
display(bayes_conditional(df_grouped, "Пол"))

display(Markdown("### Вероятность стенокардии при нагрузке по формуле Байеса"))
display(bayes_conditional(df_grouped, "Стенокардия при нагрузке"))

display(Markdown("### Вероятность возраста по формуле Байеса"))
display(bayes_conditional(df_grouped, "Возраст"))

for col in ["Возраст", "Артериальное давление", "Холестерин", "Макс. ЧСС"]:
    plt.figure(figsize=(6, 4))
    sns.barplot(
        x=col, y="Болезнь сердца", data=df_grouped, estimator=np.mean, errorbar=None
    )
    plt.title(f"Вероятность болезни в зависимости от {col}")
    plt.show()

display(
    Markdown(
        "### На барплотах видно, как изменяется вероятность болезни сердца в зависимости от значения признака."
    )
)

display(
    Markdown(
        "### Корреляционная матрица показывает линейные зависимости между признаками."
        "Например, возраст и болезнь сердца коррелируют положительно, а Макс. ЧСС коррелирует отрицательно."
    )
)

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Корреляции признаков")
plt.show()

display(
    Markdown(
        """### Из таблицы и графиков видно:

- Мужчины имеют более высокую вероятность болезни, чем женщины.

- Риск увеличивается с возрастом (особенно после 55).

- Высокий уровень холестерина и низкий ЧСС связаны с большей вероятностью болезни.

- Наличие стенокардии при нагрузке резко повышает риск."""
    )
)
```
